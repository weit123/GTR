```python
import matplotlib.pyplot as plt
import gymnasium as gym
import gym_cards
```

```python
env = gym.make('gym_cards/NumberLine-v0')
obs, info = env.reset()
plt.imshow(obs)
```

```python
obs, reward, terminated, truncated, info = env.step(1)
plt.imshow(obs)
print(reward, terminated, truncated, info)
```
